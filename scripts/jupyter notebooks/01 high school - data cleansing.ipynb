Now that I'm aware the columns don't match between files, I wrote two functions that will compare the column labels between them and create a table so I can check in detail the differences.

```python
def read_csv_files(folder_path):    
    file_paths = glob.glob(folder_path + "*.csv")    
    dfs = []    
    for file_path in file_paths:
        try:
            df = pd.read_csv(file_path, sep='|', nrows=1)
        except:
            df = pd.read_csv(file_path, sep=',', encoding='latin-1', nrows=1, usecols=range(1019))                    
        dfs.append(df)    
    combined_df = pd.concat(dfs, ignore_index=True)    
    return combined_df

def get_columns_info(folder_path):    
    df = read_csv_files(folder_path)
    columns_info = df.columns.unique()
    df_columns_info = pd.DataFrame(columns_info, columns=["Column"])

    for file_path in glob.glob(folder_path + "*.csv"):
        file_name = file_path.split("\\")[-1]
        try:
            file_df = pd.read_csv(file_path, sep='|', nrows=1)
        except:
            file_df = pd.read_csv(file_path, sep=',', encoding='latin-1', nrows=1, usecols=range(1019))
        
        file_columns = file_df.columns
        df_columns_info[file_name] = df_columns_info["Column"].isin(file_columns)    
    return df_columns_info

folder_path = "../../data/High School/Datasets/"
columns_info_df = get_columns_info(folder_path)

excel_file_path = "../../results/data/high_school_column_labels_comparison.xlsx"
columns_info_df.to_excel(excel_file_path, index=False)
```

```python
# Merge all CSV files into one df and export it as CSV file
def concatenate_csv_files(folder_path):
    file_paths = glob.glob(folder_path + "/*.csv")
    dfs = []
    
    for file_path in file_paths:
        try:
            df = pd.read_csv(file_path, sep='|')
        except:
            df = pd.read_csv(file_path, sep=',', encoding='latin-1', usecols=range(1019))
        dfs.append(df)
        
    combined_df = pd.concat(dfs, ignore_index=True)    
    return combined_df

folder_path = "../../data/High School/Datasets/"
full_data_df = concatenate_csv_files(folder_path)
print(full_data_df.shape)

excel_file_path = "../../results/data/high_school_dataset.csv"
full_data_df.to_csv(excel_file_path, index=False)
```

After reviewing the values of columns 'SUBSISTEMA_1', 'SUBSISTEMA_2' and 'SUBSISTEMA_3', they seem to be constant within each educational institution, so I wrote a script to fill in the missing data using the ffill() function

```python
full_data_df[['CCT_INS_PLA', 'PERIODO', 'SUBSISTEMA_1', 'SUBSISTEMA_2', 'SUBSISTEMA_3']].head(40)
```

```python
def fill_missing_data(data, variables):
    for variable in variables:
        data[variable] = data.groupby('CCT_INS_PLA')[variable].transform(lambda x: x.ffill())
    return data

full_data_df = fill_missing_data(full_data_df, ['SUBSISTEMA_1', 'SUBSISTEMA_2', 'SUBSISTEMA_3'])
```

In order to assign a more suitable data type to the DataFrame columns, I created a file with the DataFrame columns and their corresponding data type for each variable

```python
data_types_df = pd.read_excel('../../other_files/high_school_variables_mapping.xlsx')

for idx, data_type in data_types_df.iterrows():    
    try:
        column = data_type.values[0]
        datatype = data_type.values[1]
        full_data_df[column] = full_data_df[column].astype(datatype)
    except KeyError:
        print(f"column {column} not found in dataframe")
    except ValueError:        
        full_data_df[column].fillna(0, inplace=True)
        full_data_df[column] = full_data_df[column].astype(datatype)
    except Exception as e:    
        print(f"An error occurred with {column}:", e, type(e))
```
